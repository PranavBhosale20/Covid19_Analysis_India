```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
df = pd.read_csv("sample_covid_india.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
latest_date = df['Date'].max()
latest_data = df[df['Date'] == latest_date]

top_states = latest_data.sort_values(by='Confirmed', ascending=False)
print(top_states[['State', 'Confirmed']])
```

```python
sns.barplot(x='Confirmed', y='State', data=top_states, palette='Reds_r')
plt.title('Confirmed COVID-19 Cases by State (as of latest date)')
plt.xlabel('Confirmed Cases')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
maha_data = df[df['State'] == 'Maharashtra']

plt.plot(maha_data['Date'], maha_data['Confirmed'], label='Confirmed')
plt.plot(maha_data['Date'], maha_data['Cured'], label='Cured')
plt.plot(maha_data['Date'], maha_data['Deaths'], label='Deaths')

plt.title('COVID-19 Trend in Maharashtra')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.grid(True)
plt.show()
```
